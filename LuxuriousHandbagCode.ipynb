```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
url="https://www.amazon.in/s?k=luxury+bags+for+women&crid=29SODOTNEH2DU&sprefix=luxury+bags%2Caps%2C731&ref=nb_sb_ss_ts-doa-p_1_11"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.google.com",
    "Connection": "keep-alive",
}

r= requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
from bs4 import BeautifulSoup  # Ensure you import BeautifulSoup
import pandas as pd

Dict = {'Title': [], 'Price': [], 'Description': []}

# Extract Titles
Brands = soup.select("span.a-size-base-plus.a-color-base")
for span in Brands:
    if span and span.string:  # Ensure span is not None
        Dict["Title"].append(span.string.strip())

# Extract Prices
Prices = soup.select("span.a-price span.a-offscreen")  # More specific selection
extracted_prices = [price.get_text(strip=True) 
                    for price in Prices 
                    if price.get_text(strip=True)]

# Ensure Price List Matches Title List
while len(extracted_prices) < len(Dict["Title"]):
    extracted_prices.append("N/A")
Dict["Price"] = extracted_prices[:len(Dict["Title"])]  # Ensure alignment with titles

# Extract Descriptions
Des_Tag = soup.select("h2.a-size-base-plus.a-spacing-none.a-color-base")
for des in Des_Tag:
    if des and des.get_text(strip=True):  # Ensure text is valid
        Dict["Description"].append(des.get_text(strip=True))
    else:
        Dict["Description"].append("N/A")  # Handle missing descriptions
        
# Ensure All Lists Have Equal Length
max_len = max(len(Dict["Title"]), len(Dict["Price"]), len(Dict["Description"]))
for key in Dict:
    while len(Dict[key]) < max_len:
        Dict[key].append("N/A")

# Convert to DataFrame
df = pd.DataFrame(Dict)
print(df)
```
